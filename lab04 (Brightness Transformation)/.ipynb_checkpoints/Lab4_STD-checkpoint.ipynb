```python
import skimage.io as io
import matplotlib.pyplot as plt
import numpy as np

# Show the figures / plots inside the notebook
%matplotlib inline
from skimage.color import rgb2gray

def show_images(images,titles=None):
    #This function is used to show image(s) with titles by sending an array of images and an array of associated titles.
    # images[0] will be drawn with the title titles[0] if exists
    # You aren't required to understand this function, use it as-is.
    n_ims = len(images)
    if titles is None: titles = ['(%d)' % i for i in range(1,n_ims + 1)]
    fig = plt.figure()
    n = 1
    for image,title in zip(images,titles):
        a = fig.add_subplot(1,n_ims,n)
        if image.ndim == 2: 
            plt.gray()
        plt.imshow(image)
        a.set_title(title)
        n += 1
    fig.set_size_inches(np.array(fig.get_size_inches()) * n_ims)
    plt.show()
```

```python
'''
1
Negative Transformation
'''
def Negative(img):
    negImg = 255 - img;
    show_images([img, negImg], ['Original', 'Negative'])

image = io.imread('imgs/Picture1.png')
gray_image = rgb2gray(image[:,:,:3])
Negative(gray_image)
```

```python
'''
2
Contrast Enhancement
'''
def Contrast_enhancement(image):
    uint8Img = np.uint8(rgb2gray(image[:,:,:3])*255)
#     print(uint8Img)
#     uint8Img[((uint8Img >= 0) * (uint8Img <= 100))] = uint8Img
    m, n = uint8Img.shape
    for i in range(m):
        for j in range(n):
            if uint8Img[i][j] >= 0 and uint8Img[i][j] <= 100:
                uint8Img[i][j] = 50 + (200-50)/100
    show_images([image, uint8Img], ['Original', 'Edited'])

image = io.imread('imgs/Picture2.png')
Contrast_enhancement(image)
```

```python
'''
3
Gamma Correction
'''
def Gamma_Correction(img, c, gamma):
    edited_img = c * (img**gamma)
    show_images([image, edited_img], ['Original', 'Corrected'])

image = rgb2gray(io.imread('imgs/Picture2.png')[:,:,:3])

Gamma_Correction(image, 1, 1)

Gamma_Correction(image, 1, 3)

Gamma_Correction(image, 1, 0.5)
```

```python
'''
4 Histogram Eq.
Note: Histogram function of skimage returns only present intensity values not all 255
You can use showHist function from commonfunctions file
'''

def getImageWithHist(name,ext,nbins=256):
    image = np.uint8(rgb2gray(io.imread(name + '.' + ext)[:,:,:3])*255)
    H = np.zeros(nbins)
    
    m, n = image.shape
    for i in range(m):
        for j in range(n):
            H[image[i][j]] += 1
    H_c = [sum(H[:i+1]) for i in range(len(H))]
    q = 255*np.array(H_c)/(m*n)
    
    edited_image = np.zeros(image.shape)
    
    for i in range(0, m):
        for j in range(0, n):
            edited_image[i, j] = q[image[i, j]]
    show_images([image, edited_image], ['Before', 'After'])
getImageWithHist('imgs/Picture2', 'png')
```

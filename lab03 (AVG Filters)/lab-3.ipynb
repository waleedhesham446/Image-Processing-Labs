```python
from commonfunctions import *
import math
import skimage.io as io
import matplotlib.pyplot as plt
import numpy as np
from skimage.color import rgb2gray
from skimage.filters import median, gaussian

%matplotlib inline
%load_ext autoreload
%autoreload 2
```

```python
'''
(1) Your own median filter algorithm
'''
def Median_Filter(img):
    edited_img = np.zeros(img.shape)
    n, m = img.shape
    for i in range(n):
        for j in range(m):
            x1 = i-1
            x2 = i+2
            y1 = j-1
            y2 = j+2
            if(i == 0):
                x1 = i
            elif(i == n-1):
                x2 = i+1
            if(j == 0):
                y1 = j
            elif(j == m-1):
                y2 = j+1
            window = img[x1:x2, y1:y2]
            median = np.median(np.array(window).flatten())
            edited_img[i][j] = median
    show_images([img, edited_img], ['Before', 'After'])

image = io.imread('images/tire.tif')
noisedImg = random_noise(image = image, mode = 's&p', amount = 0.05)
Median_Filter(noisedImg)
```

```python
'''
(2) skimage median filter
'''
edited_image = median(noisedImg)
show_images([edited_image])
```

```python
'''
(3) Gaussion Filters
'''
gaussian_1 = gaussian(noisedImg, sigma = 8)
gaussian_2 = gaussian(noisedImg, sigma = 3)
gaussian_3 = gaussian(noisedImg, sigma = 0.5)

gaussian_4 = gaussian(noisedImg, sigma = 1.5)

show_images([gaussian_1, gaussian_2, gaussian_3])

io.imshow(gaussian_4)
io.show()
```
